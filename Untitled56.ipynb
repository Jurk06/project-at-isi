```python
def percentage_nan(data):
  cnt=0
  value=[]
  for i in range(0, len(data)):
    if i not in value:
      cnt=cnt+1
  c=0
  for j in range(0, len(data)):
    if str(j)=='nan':
      c=c+1
  
  print("% NAN vaue is {}".format(round((c/len(data)*100),5)))
  print("% Unique vaue is {}".format(round((cnt/len(data)*100),2)))


```

```python
def hitandkde(x, y):
    global data
    data=data.copy()
    
    plt.figure(figsize=(25,20))
    plt.subplot(3,1,1)
    sns.kdeplot(data=data, x=y, hue=x, shade=True)
    plt.title(y.upper())
    plt.xlabel(x.upper())
    plt.ylabel(y.upper())
    
    plt.subplot(3,1,2)
    sns.histplot(data=data, x=y, hue=x)
    plt.xlabel(x.upper())
    plt.ylabel(y.upper())
    
    plt.subplot(3,1,3)
    sns.histplot(data=data, x=y, hue=x, log_scale=10)
    plt.xlabel(x.upper())
    plt.ylabel(y.upper())
    plt.show()
```

```python
def trends_analysis(data, c1,c2,c3,c4,c5):
    plt.figure(figsize=(20,20))
    plt.subplot(5,1,1)
    data[c1].head(100).plot(kind='line')
    data[c1].head(100).rolling(7).mean().plot(kind='line')
    plt.title("Trends of Water level-" +c1)

    plt.figure(figsize=(20,20))
    plt.subplot(5,1,2)
    data[c2].head(100).plot(kind='line')
    data[c2].head(100).rolling(7).mean().plot(kind='line')
    plt.title("Trends of Water level-" +c2)

    plt.figure(figsize=(20,20))
    plt.subplot(5,1,3)
    data[c3].head(100).plot(kind='line')
    data[c3].head(100).rolling(7).mean().plot(kind='line')
    plt.title("Trends of Water level-" +c3)

    plt.figure(figsize=(20,20))
    plt.subplot(5,1,4)
    data[c4].head(100).plot(kind='line')
    data[c4].head(100).rolling(7).mean().plot(kind='line')
    plt.title("Trends of Water level-" +c4)

    plt.figure(figsize=(20,20))
    plt.subplot(5,1,5)
    data[c5].head(100).plot(kind='line')
    data[c5].head(100).rolling(7).mean().plot(kind='line')
    plt.title("Trends of Water level-" +c5)
    plt.show()
```

```python
trends_analysis(df, "Pre_2015", "Pre_2016", "Pre_2017", "Pre_2018","Pre_2019")
trends_analysis(data, "Pre_2015", "Pre_2016", "Pre_2017", "Pre_2018","Pre_2019")

```
